### WEEK 3 Project


#### Desi Nuskha ( A Culturally-Aware Home Remedy Advisor)

1. **Use Case: "Desi Nuskha" - A Culturally-Aware Home Remedy Advisor**  

* Choice Justification:   
  I chose this use case because access to immediate, reliable, and culturally-relevant health information is a significant challenge in Pakistan. While "desi nuskay" (home remedies) are a trusted first recourse for many, advice from family and the internet is often conflicting, unverified, and can be dangerous. This chatbot bridges the gap between traditional knowledge and evidence-based, safe practices, providing triage and guidance for common, minor ailments in a accessible, low-cost format.

* Target Users:   
Primarily adults and caregivers in Pakistan and the diaspora, especially those in semi-urban or rural areas with limited immediate access to doctors, or those seeking preliminary advice before deciding to visit a clinic.

* Expected Benefits: 
Empowerment: Provides instant, actionable health guidance.
Safety: Prevents harmful practices by offering vetted advice and clear warnings.
Triage: Helps users distinguish between minor issues and situations requiring urgent professional care.
Accessibility: Functions in Urdu and English on any device with a messaging interface, requiring no download.

* Limitations:
Not a Medical Professional: It cannot diagnose, prescribe medicine, or treat serious conditions.
Knowledge Cut-off: Its knowledge is based on its training data and may not have the latest medical research.
Lack of Personalization: It does not know the user's full medical history, allergies, or current medications, so its advice is generic.
Liability: It must be used with caution and always defer to a human doctor for any serious concerns.

**2. Justification**
Why a Chatbot? This problem is ideal for a chatbot because it requires on-demand, conversational interaction. Users need to describe specific, often nuanced symptoms (e.g., "meray gale mein khujli ho rahi hai") and receive tailored advice. A static website or app lacks this interactive, Q&A nature.

* Improvement over Alternatives:   
Compared to a website (static information), it's dynamic and personalized. Compared to a full-featured app, it's lighter, faster to develop, and easier to access (e.g., via WhatsApp). Compared to human support, it's available 24/7, instantly, and for free.

* Specific Tasks:  
 The chatbot will: 1) Triage symptoms to assess severity, 2) Provide safe, evidence-based home care instructions, 3) List what to avoid, 4) Suggest supportive dietary advice, and 5) Escalate by strongly recommending seeing a doctor when necessary.

**3. LLM Selection (Groq)**  
* Selected Model: mixtral-8x7b-32768  

* Justification:  
Size & Capability: Mixtral is a Mixture-of-Experts model. While larger than 8B models, it is exceptionally capable at following complex, multi-part instructions (like our detailed system prompt) and reasoning about symptoms, which is crucial for safety. The 32k context window allows for longer, more detailed conversations.

* Language Support:   
It has excellent performance in languages other than English, making it well-suited for understanding and generating responses in Urdu and Roman Urdu.

* Speed:  
 Groq's LPU delivers incredibly fast inference speeds for this model. This is critical for a helpful user experience; waiting a long time for health advice feels frustrating and unreliable.

```python
import requests

API_KEY = "GROQ_API_KEY" 
API_URL = "https://api.groq.com/openai/v1/chat/completions"

def get_chatbot_response(user_message, conversation_history):
    """
    Sends the user's message and conversation history to the Groq API.
    Returns the chatbot's response.
    """
    system_prompt = """
    You are 'Desi Nuskha', a helpful, cautious, and respectful home remedy advisor for common, minor health issues in Pakistan. You speak primarily in Urdu not hindi words, using a warm and conversational tone.

    **CRITICAL RULES - YOU MUST ALWAYS FOLLOW THESE:**
    1.  **Triage First:** Your first job is to assess if this is a minor issue. Always ask 1-2 clarifying questions about symptoms. IF the user describes any of the following, you MUST immediately and clearly advise them to SEE A DOCTOR and STOP. Do not provide any remedies:
        - High fever (>103°F or 39.4°C), especially in children.
        - Severe, sharp, or persistent pain.
        - Difficulty breathing.
        - Symptoms in a child under 2 years old.
        - Pregnancy.
        - Symptoms lasting more than 7 days or getting worse.
        - Any injury from a fall or accident.

    2.  **Provide Safe Advice:** Only recommend well-known, evidence-based, and safe home remedies used in Pakistan (e.g., honey/lemon for cough, ginger tea for nausea, saline gargle for sore throat).

    3.  **Structure Your Response:** If the issue is minor, format your answer clearly under these headings:
        - **Nuskha (Remedy):** Give 1-2 simple, step-by-step instructions.
        - **In Se Parhaiz Karein (Avoid This):** List common but harmful or ineffective practices related to this ailment (e.g., no antibiotics without prescription, avoid excessive sugar/oil).
        - **Diet (Kiya Khayen):** Suggest simple, easy-to-eat foods ( food acording to problem ).
        - **Aakhir Mein (Disclaimer):** ALWAYS end with this: "**Yeh salah aam takleefon ke liye hai. Agar aapki halat kharab ho ya aram na mehsoos ho, to fauran doctor se raabta karein.**"

    Begin by politely asking about their symptoms.
    """

    # Construct the message list for the API
    messages = [{"role": "system", "content": system_prompt}]
    messages.extend(conversation_history)
    messages.append({"role": "user", "content": user_message})

    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {API_KEY}"
    }
    data = {
        "model": "meta-llama/llama-4-scout-17b-16e-instruct",
        "messages": messages,
        "temperature": 0.3,
        "max_tokens": 1024,
    }

    try:
        response = requests.post(API_URL, headers=headers, json=data)
        response.raise_for_status()
        return response.json()["choices"][0]["message"]["content"]
    except Exception as e:
        return "Maafarmaani, technical masla aa gaya hai. Phir koshish karein."

# Main conversation loop
print("As-Salaam-Alaikum! Main 'Desi Nuskha' hun. Aap ko kis cheez ki takleef hai? (Type 'exit' to quit)")
conversation_history = []

while True:
    user_input = input("\nAap: ").strip()

    if user_input.lower() == 'exit':
        print("Desi Nuskha: Allah Hafiz! Khush rahein.")
        break

    if not user_input:
        continue

    bot_response = get_chatbot_response(user_input, conversation_history)
    print(f"\nDesi Nuskha: {bot_response}")

    # Save context (keep only last 4 Q&A pairs)
    conversation_history.append({"role": "user", "content": user_input})
    conversation_history.append({"role": "assistant", "content": bot_response})
    if len(conversation_history) > 8:
        conversation_history = conversation_history[-8:]
```
